```python
class Vertex:
    
    def __init__(self,performer):
        self.name = performer #name of the actor
        self.adjacent = [] #name of other co-stars
        self.edge = [] #name of movie which you can think of as weight of the edge

        
    def __str__(self):
        vertex_data = self.name + " | "
                
        for ind in range(0,len(self.adjacent)):
            vertex_data = vertex_data + self.edge[ind] + " / " + self.adjacent[ind] + " | "
        
        return vertex_data
    
    def get_data(self):
        print("vertex is ")
        print(self.name)
        print("adjacent is ")
        print(self.adjacent)
        print("edge is ")
        print(self.edge)
    

def graph_builder(file_name):
    movies_file = open(file_name,'r',encoding="ISO-8859-1") #open text file containing movies data 
    
    graph = {} #represent the graph using a dictionary indexed on the performer's name (a string).
        
    for each_line in movies_file:
        #print(each_line)
        split_data = each_line.split("/") #split the string on each line according to delimiter "/" and get array of separated date
        movie = split_data[0] #first element in the actors list is actually the movie name
        split_data[len(split_data)-1] = split_data[len(split_data)-1].replace("\n","") #remove new line in last element of split_data
        #print(split_data)
        
        #Now lets start adding data in the graph. We will represent the graph using a dictionary indexed 
        #on the performer's name (a string).
        
        #parse over the actors in split_data, check if that actor already in graph, if not add in graph by:
        # 1) calling vertex class with actor name
        # 2) populating adjacent array in the object with all co-stars for the same movie
        # 3) populating edge array with the movie name
        
        for ind in range(1,len(split_data)): #start at index 1 because index zero is movie name not actor name
            #print(split_data[ind])
            actor = split_data[ind]
            
            if actor not in graph: #because keys are actors in dictionary so we are checking if this actor is already a key
                graph[actor] = Vertex(actor)#key is actor name and value is actually an object of vertex class
                #print("current actor is " + actor)
                
                #need to add all co-stars in the adjacent array of the vertex object, co-stars are all actors
                #at indices previous to current index and actors at indices forward to current index
                
                #1:ind-1 gives us previous to current indices
                #ind+1: gives us forward to current indices
                    
                all_costars = split_data[1:ind] + split_data[ind+1: ] # + operator merges the lists into one
                #print("co_stars for current actor are")
                #print(all_costars)
                
                #assign all_costars in the self.adjacent variable of the vertex object
                #in graph, each value is actually the object of vertex class
                graph[actor].adjacent = graph[actor].adjacent + all_costars
                
                #assign movie name to the self.edge list 
                #print("movie name is " + split_data[0])
                
                for costars in all_costars: #append edges equal to number of adjacent nodes to have 1:1 relation between adjacent nodes and edges
                    graph[actor].edge.append(split_data[0])  #0th element in split_data is the movie name
            
            else: #if actor is already in the dictionary
                #print("Condition when actor is already in the dictionary")
                
                #just assign costars and movie name in the object already present in the graph
                
                all_costars = split_data[1:ind] + split_data[ind+1: ] # + operator merges the lists into one
                #print("co_stars for current actor are")
                #print(all_costars)
                
                #assign all_costars in the self.adjacent variable of the vertex object
                #in graph, each value is actually the object of vertex class
                graph[actor].adjacent = graph[actor].adjacent + all_costars #whatever is present before plus
                #the new data
                
                #assign movie name to the self.edge list 
                #print("movie name is " + split_data[0])
                for costars in all_costars:
                    graph[actor].edge.append(split_data[0])  #0th element in split_data is the movie name
    
    return graph
                

# Breadth-first-search using the Vertex class.

from collections import deque

# Run bfs and find the path from the Vertex start to the Vertex goal.
# Return that path as a list of Vertex objects.  If there is no path, return None.
def bfs(start, goal):
    frontier = deque()      # the queue of vertices we're going to visit next
    frontier.append(start)  # put the start vertex in our list of those to visit
    
    backpointer = {}
    backpointer[start] = None   # the start vertex has no backpointer
    backpointer_movie = {}
    backpointer_movie[start] = None
        
    # Keep going while there's at least one visited vertex that we have not
    # explored from yet.
    while len(frontier) > 0: 
        vertex = frontier.popleft()
        
        if vertex == goal: # If we're done, retrace the path from the goal to the start.
            path = []        
            while vertex != None: # only our start should have a None backpointer along our path
                if( backpointer_movie[vertex] != None ):
                    path.append(vertex.name + " (" + backpointer_movie[vertex] + ")")
                else:
                    path.append(vertex.name)
               
                vertex = backpointer[vertex]
                #print( vertex.name, backpointer_movie[vertex] )
            return path
        
        else: # keep exploring
            i = 0;
            for neighbor in vertex.adjacent:
                movie = vertex.edge[i]
                # Visit the vertex only if it hasn't been visited yet.  It has been
                # visited if and only if it has a backpointer.
                if not graph[neighbor] in backpointer:
                    # Set up the backpointer.  This will only happen once for each vertex,
                    # since a vertex is put in the frontier and visited only once.
                    backpointer[graph[neighbor]] = vertex
                    backpointer_movie[graph[neighbor]] = movie
                    frontier.append(graph[neighbor])
                i = i + 1
                    
    # If we get here, we've run out of vertices to explore before reaching the goal.
    return None           
        

def print_graph(graph):
    for key in graph:
        #print(key)
        #dict[key].get_data()
        print(graph[key]) #for each element in dictionary/graph call goes to __str__() of class because 
        #we are trying to print the object present there


#graph = graph_builder("movies_copy2.txt")
graph = graph_builder("movies.txt")

#print("Start Printing the Graph")
#print_graph(graph)

# start = "Hanks, Tom"
# goal = "Bacon, Kevin"

start = "Winslet, Kate"
goal = "Bacon, Kevin"

# start = "DiCaprio, Leonardo"
# goal = "Winslet, Kate"

if( (start not in graph) or (goal not in graph) ):
    print( "no path found")
else:
    path = bfs(graph[start],graph[goal])
    path = path[::-1] # reverse (start --> goal)
    print(path)

    
    
    
    
        
        
        
        
        
        
        




```
